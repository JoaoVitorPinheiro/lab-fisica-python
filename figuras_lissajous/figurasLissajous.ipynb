#Atividade de Física: Figuras de Lissajous

```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
amplitude = 4 # em radianos
vel_angular = math.pi/6 # em radianos
time = np.linspace(0,31,400) # cria 400 pontos entre 0 e 31
```

```python
f = [ 0, math.pi / 4, math.pi / 2, 3*math.pi / 4, math.pi ]
f_nome = ['0','pi / 4','pi / 2','3*pi / 4','pi']
fase = np.array(f)
fase
```

```python
#x = [amplitude*math.cos(math.radians(vel_angular*i+math.pi)) for i in time]
#y =  [amplitude*math.cos(math.radians(vel_angular*i)) for i in time]
x = [amplitude*math.cos(vel_angular*i) for i in time]
y = [amplitude*math.cos(vel_angular*i+math.pi/2) for i in time]
x = np.array(x)
y = np.array(y)
```

## 1 Descobrindo a diferença de fase entre os MHS:

```python
plt.figure(figsize=(12,6))
plt.title('Figura')
plt.xlabel('tempo')
plt.ylabel('amplitude')
plt.plot(time,x,time,y)  
plt.show()
```

#### A diferença de fase é obtida a partir da comparação das duas curvas em relação a um mesmo instante 
#### Quando uma curva está em amplitude máxima, a outra está com amplitude zero, o que indica que a difença de fase corresponde a um ângulo reto

## 2 Figura resultante dos MHS:

### 2.1 Amplitudes iguais e mesma frequência angular
#### - Amplitude: 4 
#### - Frequência Angular: pi / 6

```python
j = 0
for k in fase:
  x = [amplitude*math.cos(vel_angular*i) for i in time]
  y = [amplitude*math.cos(vel_angular*i+k) for i in time]
  x = np.array(x)
  y = np.array(y) 
  plt.figure(figsize=(12,6))
  plt.title('Fase: '+f_nome[j])
  plt.xlabel('mhs x')
  plt.ylabel('mhs y')
  plt.plot(x,y)  
  plt.axis("equal")
  plt.show()
  j=j+1
```

### 2.2 Amplitudes distintas e mesma frequência angular
#### - Amplitude de x: 4 
#### - Amplitude de y: 8
#### - Frequência Angular: pi / 6

```python
j = 0
for k in fase:
  x = [amplitude*math.cos(vel_angular*i) for i in time]
  y = [2*amplitude*math.cos(vel_angular*i+k) for i in time]
  x = np.array(x)
  y = np.array(y) 
  plt.figure(figsize=(12,6))
  plt.title('Fase: '+f_nome[j])
  plt.xlabel('mhs x')
  plt.ylabel('mhs y')
  plt.plot(x,y)  
  plt.axis("equal")
  plt.show()
  j=j+1
```
